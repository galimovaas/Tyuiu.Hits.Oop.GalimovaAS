```

public class Product
{
    public string Name { get; set; }
    public decimal Price { get; set; }
    public string Manufacturer { get; set; }
    
    public string ProductCode { get; set; }
    public int StockQuantity { get; set; }
    public DateTime ProductionDate { get; set; }
    public double Weight { get; set; }

    public Product(string name, decimal price, string manufacturer, string productCode, 
                  int stockQuantity, DateTime productionDate, double weight)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
        ProductCode = productCode;
        StockQuantity = stockQuantity;
        ProductionDate = productionDate;
        Weight = weight;
    }

    public virtual string GetInfo()
    {
        return $"Название: {Name}, Цена: {Price:C}, Производитель: {Manufacturer}";
    }

    public virtual void Discount(decimal discountPercentage)
    {
        if (discountPercentage > 0 && discountPercentage <= 100)
        {
            Price -= Price * (discountPercentage / 100);
            Console.WriteLine($"Скидка {discountPercentage}% применена. Новая цена: {Price:C}");
        }
    }

    public virtual void Display()
    {
        Console.WriteLine(GetInfo());
    }

    public virtual void UpdateStock(int quantity)
    {
        StockQuantity += quantity;
        Console.WriteLine($"Запас обновлен. Текущее количество: {StockQuantity}");
    }

    public virtual bool IsInStock()
    {
        return StockQuantity > 0;
    }

    public virtual int GetProductAge()
    {
        return (DateTime.Now - ProductionDate).Days;
    }

    public virtual double CalculateShippingCost()
    {
        return Weight * 0.5; 
    }
}


public class Electronics : Product
{
    public int WarrantyPeriod { get; set; } 
    public string PowerConsumption { get; set; }
    public string Connectivity { get; set; }
    public bool IsSmartDevice { get; set; }

    public Electronics(string name, decimal price, string manufacturer, string productCode,
                      int stockQuantity, DateTime productionDate, double weight,
                      int warrantyPeriod, string powerConsumption, string connectivity, bool isSmartDevice)
        : base(name, price, manufacturer, productCode, stockQuantity, productionDate, weight)
    {
        WarrantyPeriod = warrantyPeriod;
        PowerConsumption = powerConsumption;
        Connectivity = connectivity;
        IsSmartDevice = isSmartDevice;
    }
    public override void Discount(decimal discountPercentage)
    {
        decimal adjustedDiscount = WarrantyPeriod > 24 ? 
            discountPercentage * 0.7m : discountPercentage;
        
        if (adjustedDiscount > 0 && adjustedDiscount <= 100)
        {
            decimal oldPrice = Price;
            Price -= Price * (adjustedDiscount / 100);
            Console.WriteLine($"Скидка {adjustedDiscount:F1}% применена. " +
                            $"Старая цена: {oldPrice:C}, Новая цена: {Price:C}");
        }
    }

    public override void Display()
    {
        Console.WriteLine($"ЭЛЕКТРОНИКА: {Name}");
        Console.WriteLine($"Цена: {Price:C}, Производитель: {Manufacturer}");
        Console.WriteLine($"Гарантия: {WarrantyPeriod} месяцев");
        Console.WriteLine($"Потребление: {PowerConsumption}, Подключение: {Connectivity}");
        Console.WriteLine($"Умное устройство: {(IsSmartDevice ? "Да" : "Нет")}");
    }

    public void ExtendWarranty(int additionalMonths)
    {
        WarrantyPeriod += additionalMonths;
        Console.WriteLine($"Гарантия продлена на {additionalMonths} месяцев. " +
                         $"Общий срок: {WarrantyPeriod} месяцев");
    }

    public double CalculateEnergyCost(double hoursPerDay, double costPerKwh)
    {
        double watts;
        if (double.TryParse(PowerConsumption.Replace("W", ""), out watts))
        {
            return (watts * hoursPerDay / 1000) * costPerKwh * 30; 
        }
        return 0;
    }
}

public class Clothing : Product
{

    public string Size { get; set; }
    public string Material { get; set; }
    public string Color { get; set; }
    public string CareInstructions { get; set; }

    public Clothing(string name, decimal price, string manufacturer, string productCode,
                   int stockQuantity, DateTime productionDate, double weight,
                   string size, string material, string color, string careInstructions)
        : base(name, price, manufacturer, productCode, stockQuantity, productionDate, weight)
    {
        Size = size;
        Material = material;
        Color = color;
        CareInstructions = careInstructions;
    }

    public override void Display()
    {
        Console.WriteLine($"ОДЕЖДА: {Name}");
        Console.WriteLine($"Цена: {Price:C}, Производитель: {Manufacturer}");
        Console.WriteLine($"Размер: {Size}, Материал: {Material}, Цвет: {Color}");
        Console.WriteLine($"Инструкции по уходу: {CareInstructions}");
        Console.WriteLine($"В наличии: {StockQuantity} шт.");
    }

    public override double CalculateShippingCost()
    {
       
        return 2.0;
    }

    public string GetSizeCategory()
    {
        Dictionary<string, string> sizeCategories = new Dictionary<string, string>
        {
            ["XS"] = "Очень маленький",
            ["S"] = "Маленький",
            ["M"] = "Средний",
            ["L"] = "Большой",
            ["XL"] = "Очень большой"
        };
        
        return sizeCategories.ContainsKey(Size.ToUpper()) ? 
            sizeCategories[Size.ToUpper()] : "Стандартный";
    }

    public bool IsEcoFriendly()
    {
        List<string> ecoMaterials = new List<string> { "organic cotton", "hemp", "bamboo", "linen" };
        return ecoMaterials.Contains(Material.ToLower());
    }
}

public class Books : Product
{

    public string Author { get; set; }
    public string ISBN { get; set; }
    public int PageCount { get; set; }
    public string Genre { get; set; }

    public Books(string name, decimal price, string manufacturer, string productCode,
                int stockQuantity, DateTime productionDate, double weight,
                string author, string isbn, int pageCount, string genre)
        : base(name, price, manufacturer, productCode, stockQuantity, productionDate, weight)
    {
        Author = author;
        ISBN = isbn;
        PageCount = pageCount;
        Genre = genre;
    }

    public override string GetInfo()
    {
        return $"КНИГА: {Name}\n" +
               $"Автор: {Author}\n" +
               $"Жанр: {Genre}, Страниц: {PageCount}\n" +
               $"ISBN: {ISBN}, Цена: {Price:C}";
    }

    public override void Discount(decimal discountPercentage)
    {
        decimal maxDiscount = 30;
        decimal actualDiscount = Math.Min(discountPercentage, maxDiscount);
        
        base.Discount(actualDiscount);
    }

 
    public string GetReadingTimeEstimate()
    {
        if (PageCount <= 100) return "Быстрое чтение (1-2 часа)";
        if (PageCount <= 300) return "Среднее время (3-5 часов)";
        if (PageCount <= 600) return "Длительное чтение (6-10 часов)";
        return "Очень длинное чтение (10+ часов)";
    }

    public bool IsBestsellerCandidate()
    {
        List<string> popularGenres = new List<string> { "fiction", "romance", "mystery", "self-help" };
        return popularGenres.Contains(Genre.ToLower()) && StockQuantity > 1000;
    }
}

public interface IDigitalProduct
{
    string DownloadLink { get; set; }
    long FileSize { get; set; }
    string Format { get; set; }
    void Download();
    bool CanStream();
}

public interface IReturnable
{
    int ReturnPeriod { get; set; }
    string ReturnPolicy { get; set; }
    bool CanReturn();
    decimal CalculateRestockingFee();
}

public class EBook : Books, IDigitalProduct, IReturnable
{

    public string DownloadLink { get; set; }
    public long FileSize { get; set; }
    public string Format { get; set; }
    
    public int ReturnPeriod { get; set; }
    public string ReturnPolicy { get; set; }

    public EBook(string name, decimal price, string manufacturer, string productCode,
                int stockQuantity, DateTime productionDate, double weight,
                string author, string isbn, int pageCount, string genre,
                string downloadLink, long fileSize, string format, 
                int returnPeriod, string returnPolicy)
        : base(name, price, manufacturer, productCode, stockQuantity, productionDate, 
              weight, author, isbn, pageCount, genre)
    {
        DownloadLink = downloadLink;
        FileSize = fileSize;
        Format = format;
        ReturnPeriod = returnPeriod;
        ReturnPolicy = returnPolicy;
    }

    public void Download()
    {
        Console.WriteLine($"Загрузка книги '{Name}'...");
        Console.WriteLine($"Формат: {Format}, Размер: {FileSize / 1024 / 1024} MB");
        Console.WriteLine($"Ссылка: {DownloadLink}");
    }

    public bool CanStream()
    {
        return Format.ToLower() == "epub" || Format.ToLower() == "pdf";
    }

    public bool CanReturn()
    {
        return ReturnPeriod > 0;
    }

    public decimal CalculateRestockingFee()
    {
        return Price * 0.1m; 
    }

 
    public override double CalculateShippingCost()
    {
        return 0; 
    }

    public override void UpdateStock(int quantity)
    {
        Console.WriteLine("Электронные книги имеют неограниченный запас");
    }

    public override void Display()
    {
        Console.WriteLine($"ЭЛЕКТРОННАЯ КНИГА: {Name}");
        Console.WriteLine($"Автор: {Author}, Формат: {Format}");
        Console.WriteLine($"Размер файла: {FileSize / 1024 / 1024} MB");
        Console.WriteLine($"Период возврата: {ReturnPeriod} дней");
        Console.WriteLine($"Цена: {Price:C}");
    }
}
    
        Electronics smartphone = new Electronics(
            "Смартфон Galaxy", 799.99m, "Samsung", "ELEC-001", 
            50, new DateTime(2024, 1, 15), 0.2,
            24, "15W", "WiFi, Bluetooth, 5G", true
        );

        Clothing tshirt = new Clothing(
            "Хлопковая футболка", 29.99m, "Nike", "CLOTH-001",
            100, new DateTime(2024, 2, 1), 0.3,
            "M", "Organic Cotton", "Синий", "Стирка при 30°C"
        );

        Books novel = new Books(
            "Преступление и наказание", 15.99m, "Издательство АСТ", "BOOK-001",
            200, new DateTime(2023, 12, 1), 0.5,
            "Фёдор Достоевский", "978-5-17-123456-7", 671, "Classic"
        );

        EBook ebook = new EBook(
            "Цифровая крепость", 9.99m, "Эксмо", "EBOOK-001",
            int.MaxValue, new DateTime(2024, 3, 1), 0,
            "Дэн Браун", "978-5-04-123456-7", 450, "Thriller",
            "https://store.com/download/123", 5242880, "EPUB",
            14, "Возврат в течение 14 дней"
        );


        Product[] products = new Product[] { smartphone, tshirt, novel, ebook };

        Console.WriteLine("\n=== ИНФОРМАЦИЯ О ПРОДУКТАХ ===");
        foreach (Product product in products)
        {
            product.Display();
            Console.WriteLine($"Стоимость доставки: {product.CalculateShippingCost():C}");
            Console.WriteLine($"Возраст продукта: {product.GetProductAge()} дней");
            Console.WriteLine($"В наличии: {(product.IsInStock() ? "Да" : "Нет")}");
            Console.WriteLine(new string('-', 50));
        }
        
        smartphone.ExtendWarranty(12);
        Console.WriteLine($"Стоимость энергии в месяц: " +
                         $"{smartphone.CalculateEnergyCost(3, 0.15):C}");

        Console.WriteLine($"Категория размера: {tshirt.GetSizeCategory()}");
        Console.WriteLine($"Экологичный материал: {tshirt.IsEcoFriendly()}");

        Console.WriteLine($"Время чтения: {novel.GetReadingTimeEstimate()}");
        Console.WriteLine($"Кандидат в бестселлеры: {novel.IsBestsellerCandidate()}");

        ebook.Download();
        Console.WriteLine($"Можно стримить: {ebook.CanStream()}");
        Console.WriteLine($"Можно вернуть: {ebook.CanReturn()}");
        Console.WriteLine($"Комиссия за возврат: {ebook.CalculateRestockingFee():C}");

        Console.WriteLine("\n=== ПРИМЕНЕНИЕ СКИДОК ===");
        
        smartphone.Discount(15);
        tshirt.Discount(20);
        novel.Discount(40); 
        ebook.Discount(25);

        Console.WriteLine("\n=== ОБНОВЛЕНИЕ ЗАПАСОВ ===");
        smartphone.UpdateStock(-5); 
        tshirt.UpdateStock(50);
        ebook.UpdateStock(1000);
```
