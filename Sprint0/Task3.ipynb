Создать базовый класс Product в C#, который будет представлять информацию о 
продуктах. На основе этого класса разработать 2-3 производных класса, 
демонстрирующих принципы наследования и полиморфизма. В каждом из классов 
должны быть реализованы новые атрибуты и методы, а также переопределены 
некоторые методы базового класса для демонстрации полиморфизма.


<h2 style="color:DodgerBlue">Реализация:</h2>

----

```python
using System;

public class Product
{
    public string Name { get; set; }
    public decimal Price { get; set; }
    public string Manufacturer { get; set; }

    public Product(string name, decimal price, string manufacturer)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
    }

    public virtual string GetInfo()
    {
        return $"Название: {Name}, Цена: {Price:C}, Производитель: {Manufacturer}";
    }


    public virtual void ApplyDiscount(decimal discountPercentage)
    {
        if (discountPercentage >= 0 && discountPercentage <= 100)
        {
            Price -= Price * (discountPercentage / 100);
        }
    }

 
    public virtual void DisplayInfo()
    {
        Console.WriteLine(GetInfo());
    }
}

public class Electronics : Product
{
    public int WarrantyPeriod { get; set; }

    public Electronics(string name, decimal price, string manufacturer, int warrantyPeriod)
        : base(name, price, manufacturer)
    {
        WarrantyPeriod = warrantyPeriod;
    }


    public override void ApplyDiscount(decimal discountPercentage)
    {
        if (WarrantyPeriod > 12) 
        {
            discountPercentage += 5;
        }
        
        base.ApplyDiscount(discountPercentage);
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Гарантия: {WarrantyPeriod} месяцев";
    }
}

public class Clothing : Product
{
    public string Size { get; set; }

    public Clothing(string name, decimal price, string manufacturer, string size)
        : base(name, price, manufacturer)
    {
        Size = size;
    }

    public override void DisplayInfo()
    {
        Console.WriteLine($"Одежда: {Name}, Размер: {Size}, Цена: {Price:C}, Производитель: {Manufacturer}");
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Размер: {Size}";
    }
}

public class Books : Product
{
    public string Author { get; set; }

    public Books(string name, decimal price, string manufacturer, string author)
        : base(name, price, manufacturer)
    {
        Author = author;
    }

    public override string GetInfo()
    {
        return $"Книга: {Name}, Автор: {Author}, Цена: {Price:C}, Издательство: {Manufacturer}";
    }
}

class Program
{
    static void Main()
    {
        Console.WriteLine("Демонстрация принципов наследования и полиморфизма:");
        Console.WriteLine("===================================================");

        Product[] products = new Product[]
        {
            new Electronics("Смартфон", 50000, "Samsung", 24),
            new Clothing("Футболка", 1500, "Nike", "L"),
            new Books("Война и мир", 1200, "Эксмо", "Лев Толстой"),
            new Electronics("Ноутбук", 80000, "Apple", 12)
        };

        foreach (var product in products)
        {
            Console.WriteLine("\nДо скидки:");
            product.DisplayInfo();
            
            product.ApplyDiscount(10);
            
            Console.WriteLine("После 10% скидки:");
            product.DisplayInfo();
            Console.WriteLine("-------------------");
        }

        Console.WriteLine("\nДетальная информация:");
        Console.WriteLine("=====================");
        
        var smartphone = new Electronics("Смартфон", 50000, "Samsung", 24);
        Console.WriteLine(smartphone.GetInfo());
        
        var tshirt = new Clothing("Футболка", 1500, "Nike", "L");
        tshirt.DisplayInfo();
        
        var book = new Books("Война и мир", 1200, "Эксмо", "Лев Толстой");
        Console.WriteLine(book.GetInfo());

        Console.WriteLine("\n=== Индивидуальная работа с продуктами ===");
       
        Electronics phone = new Electronics("Телефон", 30000, "Xiaomi", 18);
        Console.WriteLine("\nЭлектроника:");
        phone.DisplayInfo();
        phone.ApplyDiscount(15);
        Console.WriteLine("После 15% скидки:");
        phone.DisplayInfo();

        Clothing jacket = new Clothing("Куртка", 5000, "Zara", "M");
        Console.WriteLine("\nОдежда:");
        jacket.DisplayInfo();
        jacket.ApplyDiscount(20);
        Console.WriteLine("После 20% скидки:");
        jacket.DisplayInfo();

        Books novel = new Books("Преступление и наказание", 800, "АСТ", "Достоевский");
        Console.WriteLine("\nКниги:");
        novel.DisplayInfo();
        novel.ApplyDiscount(25);
        Console.WriteLine("После 25% скидки:");
        novel.DisplayInfo();
    }
}
```
