```polyglot-notebook
using System;


    public class Product
    {
      
        private string name;
        private decimal price;
        private int quantity;
        protected string category;
        internal string description;
        
      
        public Product(string name, decimal price, int quantity, string category, string description)
        {
            this.name = name;
            this.price = price;
            this.quantity = quantity;
            this.category = category;
            this.description = description;
        }
      
        public string Name
        {
            get { return name; }
            set { name = value; }
        }
        
        public decimal Price
        {
            get { return price; }
            set { price = value; }
        }
        
        public int Quantity
        {
            get { return quantity; }
            set { quantity = value; }
        }
        
        public string Category
        {
            get { return category; }
            set { category = value; }
        }
        
        public string Description
        {
            get { return description; }
            set { description = value; }
        }
        
        public decimal GetTotalPrice()
        {
            return price * quantity;
        }
        
        public void UpdateQuantity(int amount)
        {
            quantity += amount;
            if (quantity < 0) quantity = 0;
        }
        
        public string GetProductInfo()
        {
            return $"Товар: {name}, Цена: {price:C}, Количество: {quantity}, " +
                   $"Категория: {category}, Описание: {description}";
        }
    }
    
   
    public class Seller
    {
       
        private string name;
        private string employeeId;
        protected decimal salary;
        internal string contactInfo;
        private System.Collections.Generic.List<Product> products;
        
      
        public Seller(string name, string employeeId, decimal salary, string contactInfo)
        {
            this.name = name;
            this.employeeId = employeeId;
            this.salary = salary;
            this.contactInfo = contactInfo;
            this.products = new System.Collections.Generic.List<Product>();

        }
        public string Name
        {
            get { return name; }
            set { name = value; }
        }
        
        public string EmployeeId
        {
            get { return employeeId; }
            set { employeeId = value; }
        }
        
        public decimal Salary
        {
            get { return salary; }
            set { salary = value; }
        }
        
        public string ContactInfo
        {
            get { return contactInfo; }
            set { contactInfo = value; }
        }
    
        public void AddProduct(Product product)
        {
            products.Add(product);
            Console.WriteLine($"Товар '{product.Name}' добавлен продавцу {name}");
        }
        
        public void SellProduct(Product product, int quantity)
        {
            if (product.Quantity >= quantity)
            {
                product.UpdateQuantity(-quantity);
                Console.WriteLine($"Продано {quantity} единиц товара '{product.Name}'");
            }
            else
            {
                Console.WriteLine($"Недостаточно товара '{product.Name}' на складе");
            }
        }
        
        public string GetSellerInfo()
        {
            return $"Продавец: {name}, ID: {employeeId}, Зарплата: {salary:C}, Контакты: {contactInfo}";
        }
        
        public System.Collections.Generic.List<Product> GetProducts()
        {
            return products;
        }
    }

    public class Store
    {
        private string storeName;
        private string location;
        public string storeHours;
        private System.Collections.Generic.List<Seller> sellers;
        private System.Collections.Generic.List<Product> products;
  
        public Store(string storeName, string location, string storeHours)
        {
            this.storeName = storeName;
            this.location = location;
            this.storeHours = storeHours;
            this.sellers = new System.Collections.Generic.List<Seller>();
            this.products = new System.Collections.Generic.List<Product>();
        }
      
        public string StoreName
        {
            get { return storeName; }
            set { storeName = value; }
        }
        
        public string Location
        {
            get { return location; }
            set { location = value; }
        }
        
        public string StoreHours
        {
            get { return storeHours; }
            set { storeHours = value; }
        }
        
        
        public void AddSeller(Seller seller)
        {
            sellers.Add(seller);
            Console.WriteLine($"Продавец {seller.Name} добавлен в магазин {storeName}");
        }
        
        public void AddProduct(Product product)
        {
            products.Add(product);
        }
        
        public void ListProducts()
        {
            Console.WriteLine($"\nТовары в магазине '{storeName}':");
            foreach (var product in products)
            {
                Console.WriteLine(product.GetProductInfo());
            }
        }
        
        public string GetStoreInfo()
        {
            return $"Магазин: {storeName}, Адрес: {location}, Часы работы: {storeHours}, " +
                   $"Количество продавцов: {sellers.Count}, Количество товаров: {products.Count}";
        }
        
        public System.Collections.Generic.List<Seller> GetSellers()
        {
            return sellers;
        }
    }
    
            Product product1 = new Product("Ноутбук", 60000m, 10, "Электроника", "Мощный ноутбук");
            
            Seller seller1 = new Seller("Тимофей Р.", "EMP001", 40000m, "tima7@store.com");
            
            Store store1 = new Store("ДНС", "ул. Центральная, 1", "9:00-21:00");
            
          
            store1.AddProduct(product1);
            store1.AddSeller(seller1);
            seller1.AddProduct(product1);
            
            Console.WriteLine("Информация о товаре");
            Console.WriteLine(product1.GetProductInfo());
            Console.WriteLine($"Общая стоимость: {product1.GetTotalPrice():C}");
            
            Console.WriteLine("Информация о продавце");
            Console.WriteLine(seller1.GetSellerInfo());
            
            Console.WriteLine(" Информация о магазине ");
            Console.WriteLine(store1.GetStoreInfo());
            
            Console.WriteLine(" Продажа товара");
            seller1.SellProduct(product1, 2);
            
            Console.WriteLine(" Обновленная информация о товаре ");
            Console.WriteLine(product1.GetProductInfo());
            
            Console.WriteLine(" Список товаров в магазине");
            store1.ListProducts();
            
            Console.WriteLine(" Обновление количества товара");
            product1.UpdateQuantity(5);
            Console.WriteLine($"Новое количество: {product1.Quantity}");
            
            Console.WriteLine("Демонстрация работы геттеров и сеттеров ");
            product1.Price = 30000m;
            seller1.Salary = 80000m;
            store1.StoreHours = "10:00-21:00";
            
            Console.WriteLine($"Новая цена товара: {product1.Price:C}");
            Console.WriteLine($"Новая зарплата продавца: {seller1.Salary:C}");
            Console.WriteLine($"Новые часы работы магазина: {store1.StoreHours}");
```
